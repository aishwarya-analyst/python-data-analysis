```python
days=[1,2,3,4,5,6,7]
max_t=[62,64,66,69,68,61,63]
min_t=[56,54,57,58,51,52,59]
avg_t=[33,34,32,36,38,39,31]
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend(loc="best",shadow=True)
plt.grid()           
```

```python
#BAR CHART
company=['A','B','C','D']
revenue=[90,115,56,89]
plt.bar(company,revenue)
```

```python
import numpy as np
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
```

```python
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('US Tech stocks')
plt.bar(xpos,revenue,label="revenue")
plt.legend()
```

```python
profit=[10,20,40,30]
plt.bar(xpos,profit,label="profit")
plt.legend()
```

```python
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('US Tech stocks')
plt.bar(xpos,revenue,label="revenue")
profit=[10,20,40,30]
plt.bar(xpos,profit,label="profit")
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue,label="revenue")
plt.bar(xpos+0.2,profit,label='profit')
plt.legend()
```

```python
#HISTOGRAMS
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.5)
```

```python
plt.xlabel('sugar_levels')
plt.ylabel('number of persons')
plt.hist(blood_sugar,bins=[80,100,125,150])
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','red'],label=['men','women'])
plt.legend()
```
